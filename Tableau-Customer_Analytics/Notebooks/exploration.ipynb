## Recency, Frequenc, Monetary analysis (RFM)

```python
reference_date = sales_dataset["order_purchase_timestamp"].max()
print(reference_date)
```

```python
rfm = sales_dataset.groupby("customer_unique_id").agg(
    {
        "order_purchase_timestamp": "max",
        "order_id": "nunique",
        "revenue":"sum"
    }
)
rfm.columns = ["last_purchase","frequency","monetary"]
rfm
```

```python
rfm["recency"] = (reference_date - rfm["last_purchase"]).dt.days
rfm =rfm[["recency", "frequency", "monetary"]] 
```

```python
rfm.head()
```

**Recency scoring**

```python
rfm["r_score"] = pd.qcut(rfm["recency"],5,labels=[5,4,3,2,1])
rfm[["recency","r_score"]] 
```

**Frequency scoring**

```python
rfm["f_score"] = rfm["frequency"].apply(lambda x:
    1 if x == 1 else
    2 if x == 2 else
    3 if x == 3 else
    4 if x == 4 else
    5
)
rfm[["frequency","f_score"]]
```

```python
rfm["frequency"].value_counts().head(10)
```

```python
rfm["f_score"].value_counts().head(10)
```

**Monetary scoring**

```python
rfm["m_score"] = pd.qcut(rfm["monetary"], 5, labels=[1,2,3,4,5])
rfm[["monetary","m_score"]]
```

**RFM scoring**

```python
rfm["rfm_score"]= (rfm["r_score"].astype(str) + rfm["f_score"].astype(str) + rfm["m_score"].astype(str))
rfm
```

```python
rfm["frequency"].max()
```

**Segment function**

```python
def segment_customer(row):

    r = int(row["r_score"])
    f = int(row["f_score"])

    if r >= 4 and f >= 4:
        return "Champions" #recent and very frequent

    elif r >= 3 and f >= 3:
        return "Loyal Customers" #recent and frequent

    elif r >= 4 and f <= 2:
        return "Potential Loyalists" #recent and low frequency

    elif r <= 2 and f >= 3:    
        return "At Risk"  #old and mid frequency

    else:
        return "Lost"       
```

```python
rfm["segment"] = rfm.apply(segment_customer,axis =1)
rfm["segment"].value_counts()
```

**Exporting table**

```python
rfm.reset_index(inplace=True)
#rfm.to_csv("rfm_segmentation.csv", index=False)
```

#Cohor Retention table creation#

```python
sales_dataset["order_month"] = sales_dataset["order_purchase_timestamp"].dt.to_period("M")
first_purchase =sales_dataset.groupby("customer_unique_id")["order_month"].min()
```

```python
sales_dataset = sales_dataset.merge(first_purchase, on="customer_unique_id", suffixes=("","_cohort"))
sales_dataset
```

**Client age (cohort index)**

```python
sales_dataset["cohort_index"] = (sales_dataset["order_month"].astype(int) - sales_dataset["order_month_cohort"].astype(int))
sales_dataset["cohort_index"].value_counts()
```

**Calculate the cohort table**

```python
cohort_data = sales_dataset.groupby(
    ["order_month_cohort", "cohort_index"]
)["customer_unique_id"].nunique().reset_index()
```

```python
cohort_data
```

**Transform the data in matrix (pivot)**

```python
cohort_table = cohort_data.pivot(
    index="order_month_cohort",
    columns="cohort_index",
    values="customer_unique_id"
)
cohort_table
```

**Retention rate**

```python
cohort_size = cohort_table.iloc[:,0]
retention_table = cohort_table.divide(cohort_size, axis=0) *100
retention_table = retention_table.reindex(columns= range(0,21))
retention_table
```

## Revenue per cohort table

```python
revenu_cohort = sales_dataset.groupby(["order_month_cohort", "cohort_index"])["revenue"].sum().reset_index()
revenu_cohort
```

**Transform the data in matrix (pivot)**

```python
revenue_cohort_table = revenu_cohort.pivot(index="order_month_cohort", columns="cohort_index",values = "revenue")
revenue_cohort_table
```

```python
cohort_revenue_base = revenue_cohort_table.iloc[:,0]

revenue_retention = revenue_cohort_table.divide(cohort_revenue_base, axis=0) * 100
revenue_retention = revenue_retention.reindex(columns=range(0,21))
revenue_retention
```

### Export the retention table

```python
revenue_retention.reset_index(inplace=True)

#revenue_retention.to_csv("revenue_retention_cohort.csv", index=False)
```
